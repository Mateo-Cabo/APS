## Punto a)

Para hallar la funcin $$H = \frac{V_{o}}{V_{i}}$$ se procedi de la siguiente forma:

$$H = \frac{Z_{R}}{Z_{C}+Z_{L}+Z_{R}}$$

Sabiendo que $$Z_{R} = R, \quad Z_{C}(s) = \frac{1}{sC}, \quad Z_{L}(s) = sL$$

$$H(s) = \frac{R}{\frac{1}{sC}+sL+R}$$

$$H(s) = \frac{R}{\frac{1}{sC}+sL+R}*\frac{sC}{sC}$$

$$H(s) = \frac{RsC}{1+s^{2}LC+RsC}$$

$$H(s) = \frac{s\frac{R}{L}}{\frac{1}{LC}+s^{2}+s\frac{R}{L}}*\frac{LC}{LC}$$

$$H(s) = \frac{s\frac{\omega_{0}}{Q}}{s^{2}+s\frac{\omega_{0}}{Q}+\omega_{0}^{2}}$$

$$H(j\omega) = \frac{j\omega\frac{\omega_{0}}{Q}}{(j\omega)^{2}+j\omega\frac{\omega_{0}}{Q}+\omega_{0}^{2}}$$

$$\left | H(j\omega) \right | = \frac{\omega\frac{\omega_{0}}{Q}}{\sqrt{(\omega_{0}^{2}-\omega^{2})^{2}+\omega^{2}(\frac{\omega_{0}}{Q})^2}}$$

$$\displaystyle \lim_{ \omega \to 0} \left | H(j\omega) \right | \quad \Rightarrow \quad \omega^{2}$$

$$\displaystyle \lim_{ \omega \to \infty} \left | H(j\omega) \right | \quad \Rightarrow \quad \frac{1}{\omega^{2}}$$

## Punto b)

Procediendo de forma similar al caso anterior

$$H = \frac{Y_{R}+Y_{L}}{Y_{R}+Y_{L}+Z_{C}}$$

$$H = \frac{(\frac{1}{Z_{R}}+\frac{1}{Z_{L}})^{-1}}{(\frac{1}{Z_{R}}+\frac{1}{Z_{L}})^{-1}+Z_{C}}$$

$$H(s) = \frac{(\frac{1}{R}+\frac{1}{sL})^{-1}}{(\frac{1}{R}+\frac{1}{sL})^{-1}+\frac{1}{sC}}$$

$$H(s) = \frac{\frac{sLR}{R + sL}}{\frac{sLR}{R + sL}+\frac{1}{sC}}$$

$$H(s) = \frac{R+sL}{R+sL}*\frac{sLR}{sLR+\frac{R+sL}{sC}}$$

$$H(s) = \frac{sC}{sC}*\frac{sLR}{sLR+\frac{R+sL}{sC}}$$

$$H(s) = \frac{s^{2}LRC}{s^{2}LRC+R+sL}$$

$$H(s) = \frac{LC}{LC}*\frac{s^{2}R}{s^{2}R+\frac{R}{LC}+\frac{s}{C}}$$

$$H(s) = \frac{R}{R}*\frac{s^{2}}{s^{2}+\frac{1}{LC}+s\frac{1}{RC}}$$

$$H(s) = \frac{s^{2}}{s^{2}+\omega_{0}^{2}+s\frac{1}{RC}}$$

$$H(j\omega) = \frac{(j\omega)^{2}}{(j\omega)^{2}+\omega_{0}^{2}+j\omega\frac{1}{RC}}$$

$$\left | H(j\omega) \right | = \frac{\omega^{2}}{\sqrt{(\omega_{0}^{2}-\omega^{2})^{2}+\omega^{2}(\frac{1}{RC})^2}}$$

$$\displaystyle \lim_{ \omega \to \infty} \left | H(j\omega) \right | \quad \Rightarrow \quad \omega^{2}$$
