```python
from IPython.display import Image

from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
import scipy
import math
import scipy.optimize
```

## Punto a)

```python
Image(filename='TS6-a-1.png') 
```

```python
fs = 4000

w = 2*np.pi * fs

Res = np.pi/fs

omega = np.arange(0, 2*np.pi, Res)
omega2 = np.arange(0, np.pi, Res)

arg = omega  * 1j

z = np.exp(arg)

H = z**0 + z**-1 + z**-2 + z**-3
H_mod = abs(H)

ceros = []
for x in range(len(H_mod)):
    if H_mod[x] < 0.000001:
        ceros.append(x)

modulo = np.abs(H)

fase = np.angle(H)

plt.figure(figsize=(20,10))
plt.plot(omega2,abs(H[0:fs]))
plt.title('Respuesta de mdulo')
plt.ylabel('Modulo de H')
plt.xlabel('Frecuencia Omega')
plt.show

plt.figure(figsize=(10,10))
plt.plot(np.cos(omega), np.sin(omega))
for x in ceros:
    plt.plot(np.cos(x*Res), np.sin(x*Res), marker = 'o', color = 'red', label = 'Polo')
plt.title('Diagrama de polos y ceros')
plt.plot(0,0, marker = 'x', color = 'green', label = 'Ceros')
plt.legend()
plt.show

plt.figure(figsize=(20,10))
plt.title('Respuesta de fase')
plt.ylabel('Fase de H')
plt.xlabel('Frecuencia Omega')
plt.axhline(y = -np.pi/2, linestyle = 'dashed', color = 'red', label = '-/2')
plt.plot(omega2, fase[0:fs])
plt.legend()
plt.show
```
